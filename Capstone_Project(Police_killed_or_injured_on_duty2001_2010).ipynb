```python
# Import Statements

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("02_01_District_wise_crimes_committed_against_SC_2001_2012.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
plt.bar(df['Year'],df['Other Crimes Against SCs'])  # Trying to observe the crime rate in India over the years
plt.title('Bar chart of Crime rate in India between 2001 & 2010')
plt.xlabel('Year')
plt.ylabel('Number of Victims')
plt.show()
```

From the above graph we can say that the highest crime rate in India is in 2002 & Lowest in 2005.

```python
df = df.sort_values('Other Crimes Against SCs') #Sorting the dataset in the ascending order of total number of crimes
df.head()
```

```python
df.plot(figsize=(20, 10))

plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df['Year'], df['Other Crimes Against SCs'])
plt.title("Other Crimes Against SCs")
plt.show()
```
