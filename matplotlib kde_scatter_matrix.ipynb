```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


from pandas.plotting  import scatter_matrix
```

```python
iris = sns.load_dataset('iris')
df = iris.drop(['species'],axis = 1)
df.head()
```

```python
# kde Kernel density estimition

scatter_matrix (df, figsize =(10,8),color ='g',diagonal = 'kde')
plt.show()
```

```python
scatter_matrix (df, figsize =(10,8),color ='b',diagonal = 'bar')
plt.show()
```

```python
scatter_matrix (df, figsize =(10,8),color ='r',diagonal = 'hist')
plt.show()
```

```python
scatter_matrix (df, figsize =(10,8),color ='g',diagonal = '%hist')
plt.show()
```
