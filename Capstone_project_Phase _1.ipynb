```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
```

```python
url = 'https://www.indiacensus.net/literacy-rate.php'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
state = []     # empty list for store the 


for i in soup.find_all('a'):
    state.append(i.text)
```

```python
state
```

```python
populations=[]
for i in soup.find_all('div',class_="txt-right"):
    populations.append(i.text.split('|')[0])
    
populations
```

```python
literacy=[]
for i in soup.find_all('b'):
    literacy.append(i.text.split('|')[0])
    
literacy
```

```python
data=pd.DataFrame()
data['State'] = state[6:42]
data['Population'] = populations[0:36]
data['Literacy Rate'] = literacy[0:36]
data
```
